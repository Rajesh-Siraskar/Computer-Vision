## Create Histogram of Gradients (HoG) as feature vectors

**Parameters**:

- cell_size: size of cell in nxn pixels (default (8, 8))
- block_cells: multiplier i.e. m cells high and wide (default=2) 
- step_cells: number of cells by which a step is made while scanning bounding box left to right and top to bottom (default=1)
- num_bins: number of 'directional' (bins of direction i.e. angle of the gradient vectors) (default = 9)

```python
n_images = images.shape[0]
n_image_height = images.shape[1]
n_image_width = images.shape[2]

image_size = (n_image_height, n_image_width)

# Cell size of 8 pixels x 8 pixels
cell_size = (8, 8)

# Block size of 2x2 cells
block_size =   2

# Block stride - can move say one cell at a time, scanning left to right and top to bottom
block_stride = 1

# Number of 'directional' bins: 180 degs./20 deg bins = 9 bins
num_bins = 9

# Filtering Options 'BILATERAL', 'GAUSSIAN' 
# For Gaussian: All three parameters (ksize.width, ksize.height) and sigma assumed to take the value filter_param
# For Bilateral: filter_param = Sigma values
# Sigma values: Small (< 10), filter will not have much effect, if large (> 150), 
#     = > strong effect, making the image look sketch like/cartoonish
    
apply_filter = 'BILATERAL'
filter_param = 120

print('\n HoG Parameters')
print('   image_size:', image_size)
print('   cell_size:', cell_size)
print('   block_size:', block_size)
print('   block_stride:', block_stride)
print('   num_bins:', num_bins)
print('   apply_filter:', apply_filter)
print('   filter_param:', filter_param)
```

## Creating the optimised model

GridSearchCV() will look for the best parameters to create a model with. To ensure that the optimised model serves the purpose of pedestrian detection best we can set the **scoring** parameter.

Classification models are evaluated using multiple scoring metrics

**Confusion matrix**: Provides a measure of 'impurity transmission'. Non-pedestrian objects being classified as pedestrians and pedestrians being classified as back-grounds.

**Precision**: Accuracy of the positive predictions (Geron 2017: 85)
    
    Precision = TP/(TP+FP)

**Recall (a.k.a. Sensitivity or True Positive Rate)**: Ratio of positive class instances identified correctly
    
    Recall = TP/(TP+FN)

**F1 score**: Combines both metrics into a single one. Harmonic mean is used to achieve a higher score only if both precision and recall are also high

    F1 = 2 / (1/P + 1/R)

### For a Pedestrian detection system:

While a high value of TP is definately desired, for a pedestrian system it is more important to have a high **recall** than high precision. This will ensure low False Negatives i.e. missing a pedestrian and assuming it to be a non-pedestrian

#### Initiate the grid-search to obtain optimum parameters using random forests classifier

```python
%%time
grid = GridSearchCV(svm_classifier, grid_parameters, scoring='recall', cv=5)
```

#### Call fit to perform grid search and find parameters for the best model

The fit() method will fit the model for each combination of parameters (i.e. each grid _point_). Scores are maintained for each fit

```python
%%time
grid.fit(X,y)
```

### Area under the ROC curve

ROC curve plots **recall** (i.e. TPR, true positive rate) vs false positive rate (i.e. FPR, Negative instances incorrectly classified as  positive) (Geron 2017: 91)

- Ideal curve hugs the top left and larger the area the better the classifier is (James G., et al 2015: 147)
- ROC curves are the useful for comparing different classifiers (James G., et al 2015: 147)

Function below (Geron 2017: 91) helps plots a ROC curve

```python
## ROC curve plotting function (Geron 2017: 91)
def plot_roc_curve (fpr, tpr):
    
    # plot the FPR vs TPR curve
    plt.plot(fpr, tpr, linestyle = 'solid', linewidth=3, color='#49a4aa')
    
    # plot the straight line
    plt.plot([0, 1], [0, 1], linestyle = 'dotted', color='gray')
    
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC curve')
```

Compute the TPR and FPR and plot ROC curve

```python
fpr, tpr, thresholds = roc_curve(y, y_scores)

plot_roc_curve(fpr, tpr)
plt.show()
```

```python
precisions, recalls, thresholds = precision_recall_curve(y, y_scores)
average_precision = average_precision_score(y, y_scores)
AUC_score = roc_auc_score(y, y_scores)

print('Average precision-recall score: {0:0.3f}'.format(average_precision))
print('Area under the ROC curve:       {0:0.3f}'.format(AUC_score))
```

## Precision/Recall tradeoff and Decision thresholds

Shifting the **decision thresholds** causes one of the classifier properties (precision or recall) to increase and decrease the other

Finding this threshold requires plotting decision scores (instead of predictions which are just binary 0/1s). Probability scores of the classifier can be considered as decision scores.

Function below (Geron 2017: 89) helps plots such a curve

```python
## Plot Precision/Recall tradeoff vs decision thresholds (Geron 2017: 89)
def plot_precision_recall_vs_threshold (precisions, recalls, thresholds):
    plt.plot(thresholds, precisions[:-1], linestyle = 'solid', linewidth=3, color='#49a4aa', label='Precision')
    plt.plot(thresholds, recalls[:-1], linestyle = 'solid', linewidth=3, color='#b86b77', label='Recall')
    plt.xlabel('Threshold')
    plt.legend(loc='lower left')
    plt.title('Precision-Recall tradeoff vs Decision threshold')
    plt.ylim([0.50, 1.05])
```

```python
plot_precision_recall_vs_threshold(precisions, recalls, thresholds)
```

### Decision threshold = Probability for decision

From plot above indicates that the probability threshold can be pushed as high as **0.99**, only beyond this does recall suffer at the cost of precision

### Learning curve

- Identify how the **penalty** parameter determining the error term affects learning curves 
- Shows generalisation ability for unseen data

```python
## Plot learning curve for number of estimators
            
fig, ax = plt.subplots(1, 3, figsize=(16, 6))
fig.subplots_adjust(left=0.0625, right=0.95, wspace=0.1)

# Vary C
for i, C in enumerate([1.0, 10.0, 50.0]):
    N, train_lc, val_lc = learning_curve(SVC(C=C), X, y, cv=5,
                                         train_sizes=np.linspace(0.1, 1.0, 20))

    ax[i].plot(N, np.mean(train_lc, 1), color='blue', label='Training score')
    ax[i].plot(N, np.mean(val_lc, 1), color='red', label='Validation score')
    ax[i].hlines(np.mean([train_lc[-1], val_lc[-1]]), N[0], N[-1],
                 color='gray', linestyle='dashed')

    ax[i].set_ylim(0.0, 1.2)
    ax[i].set_xlim(N[0], N[-1])
    ax[i].set_xlabel('Training size')
    ax[i].set_ylabel('Score')
    ax[i].set_title('C = {0}'.format(C), size=14)
    ax[i].legend(loc='best')
```
